```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 18})
```

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
```

```python
theta1 = np.arange(-2 * np.pi, 2 * np.pi+np.pi/2, step=(np.pi / 2))
theta2 = np.arange(-2 * np.pi, 2 * np.pi+np.pi/2, step=(np.pi))
th1_ticks = ['-2π', '-3π/2', 'π', '-π/2', '0', 'π/2', 'π', '3π/2', '2π']
th2_ticks = ['-2π', 'π', '0', 'π', '2π']
```

```python
plt.figure(figsize=(8,5))
plt.plot(x,np.cos(x),label='wave 1')
plt.plot(x,np.cos(x+np.pi/4),label='wave 2')
plt.axhline(y=0, color='k', linestyle='-',linewidth=2)
plt.axvline(x=0, color='k', linestyle='-',linewidth=2)
theta = np.arange(-2 * np.pi, 2 * np.pi+np.pi/2, step=(np.pi / 2))
plt.xticks(theta1, th1_ticks)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
